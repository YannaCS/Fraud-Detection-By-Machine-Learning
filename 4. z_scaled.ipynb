```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('tmp_res_rfe.csv')
df.head()
```

```python
df['fraud_label'].sum()
```

```python
df.shape
```

```python
df.drop(df.columns[0],axis=1,inplace=True)
df.head()
```

```python
label=df.fraud_label
label.shape
```

```python
only_feature = df.drop(columns = ['fraud_label'])
std = only_feature.std()
mean = only_feature.mean()
z_scaled = (only_feature - mean)/std
```

```python
column = z_scaled.columns
z_scaled.loc[:,column] = z_scaled[column].clip(upper=1000)
z_scaled.loc[:,column] = z_scaled[column].clip(lower=-1000)
```

```python
z_scaled.describe()
```

```python
z_copy=z_scaled.copy()
z_copy['fraud_label']=label
z_copy.to_csv('D://cis//Stephen Coggeshall//z_scaled.csv',index=True,header=True)
```
